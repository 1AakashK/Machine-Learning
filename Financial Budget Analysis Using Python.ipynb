```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Importing Data

```python
data = pd.read_csv("C:/Users/aakas/Downloads/India_budget_2021.csv")
data
```

See for NaN values

Drop NaN values

```python
data.dropna()
```

Prepare the data by only selecting the main departments and putting all the other departments in the other category:

```python
data = data.iloc[[0,8,11,14,18,23,41,42,43],:]
```

Create a new row

```python
row = {'Department /Ministry': 'OTHERS', 'Fund allotted(in ₹crores)': 592971.0800000001}
```

```python
# Append the row

data = data.append(row,ignore_index= 'True')
print(data)
```

Plot

```python
data.plot.bar(x='Department /Ministry', y='Fund allotted(in ₹crores)', rot=45)
plt.xlabel('Department / Ministry')
plt.ylabel('Fund allotted (in ₹crores)')
plt.title('Ministry-wise Fund Allocation')
plt.show()
```

Plot this data into a donut plot to have a clear view of the distribution of funds among all the departments

```python
df = data["Fund allotted(in ₹crores)"]
labels = data["Department /Ministry"]
plt.figure(figsize=(8,8))
plt.pie(df, labels=labels, autopct='%1.1f%%', startangle=90, pctdistance=0.85, shadow =True)
central_circle = plt.Circle((0, 0), 0.5, color='white')
fig = plt.gcf()
fig.gca().add_artist(central_circle)
plt.rc('font', size=12)
plt.title("Distribution of The Budget", fontsize=35)
plt.show()
```
